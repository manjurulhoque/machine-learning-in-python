```python
import pandas as pd
```

```python
# define column names
names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'class']
```

```python
df = pd.read_csv('iris.data', header=None, names=names)
df.head()
```

```python
dir(df)
```

```python
# loading spliting library
from sklearn.model_selection import train_test_split
# library for accuracy
from sklearn.metrics import accuracy_score
```

```python
X = np.array(df.ix[:, 0:4])
X
```

```python
y = np.array(df['class'])
y
```

```python
# split into train and test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
classifier = KNeighborsClassifier(n_neighbors=3)
classifier.fit(X_train, y_train)
```

```python
prediction = classifier.predict(X_test)
```

```python
prediction
```

```python
# accuracy 
accuracy_score(y_test, prediction) # y_test is for calculate accuracy with y_test target value
```
